```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

import torch
import torchvision
from torchvision import transforms

import os
import matplotlib.pyplot as plt

import glob

plt.gray()

files = glob.glob(os.getcwd() + '/result_img/*')
for file in files:
    print(file)
```

```python
img_list = []
for file in files:
    img = cv2.imread(file)
    im_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)    
    im_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)    
    ret, im_threshold = cv2.threshold(im_gray, 128, 255, cv2.THRESH_BINARY)
    plt.imshow(im_threshold)
    plt.show()
    img_list.append(im_rgb)
```
